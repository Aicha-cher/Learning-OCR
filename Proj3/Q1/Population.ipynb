```python
import pandas as pd
df=pd.read_csv("FAOSTAT_data_5-22-2018 (1).csv")
#Ce fichier comporte la population totale de tous les pays pour les années 2010 2011 2012 et 2013
#La Chine comme agrégat, faisant partie de la séléction normale de tous les pays, est enlevée de la sélection
#dans ce fichier et tous les autres avec lesquels je travaillerai.
```

```python
df=df.groupby("Année").get_group(2013)
```

```python
#Population totale en 2013, en milliers de gens
df.sum()["Valeur"]
```

```python
#Pour des réutilisations futures, je vais la "nettoyer" ici
Popu2013=df.drop(columns=["Code Domaine","Domaine","Code Pays","Code Élément","Élément","Code Produit","Produit","Code Année","Année","Unité","Symbole","Description du Symbole"])
Popu2013["Population"]=Popu2013["Valeur"]*1000
Popu2013=Popu2013.drop(columns="Valeur")
Popu2013=Popu2013.reset_index().drop(columns="index")
```
