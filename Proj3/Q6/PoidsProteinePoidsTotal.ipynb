```python
import pandas as pd

data=pd.read_csv("../Q1/Popu2013.csv")
FAOSTAT_Prot=pd.read_csv("FAOSTAT_Prot.csv")

def conversionProt(valeur, population):
    return valeur*population*365

def Calculfinal(proteinedispo,nourrituredispo):
    return (proteinedispo/nourrituredispo)*100

def ConversionGr(valeur,population):
    return valeur*1000*population
```

```python
DispoAlimQte=FAOSTAT_Prot.groupby("Élément").get_group("Disponibilité alimentaire en quantité (kg/personne/an)")
```

```python
DispoProt=FAOSTAT_Prot.groupby("Élément").get_group("Disponibilité de protéines en quantité (g/personne/jour)")
```

```python
DispoProt=DispoProt.drop(columns=["Domaine","Élément","Année"])
DispoProt=DispoProt.rename(columns={"Valeur":"Protéine dispo(g/pers/jour)"})
```

```python
DispoAlimQte=DispoAlimQte.rename(columns={"Valeur":"Total dispo (kg/pers/an)"})
DispoAlimQte=DispoAlimQte.drop(columns=["Domaine","Élément","Année"])
```

```python
DispoAlimQtePopu=pd.merge(DispoAlimQte,data,on="Pays")
DispoAlimQtePopu=DispoAlimQtePopu.assign(TotGrDispoAlim=ConversionGr(DispoAlimQtePopu["Total dispo (kg/pers/an)"],DispoAlimQtePopu.Population))
TotalDispo=DispoAlimQtePopu.drop(columns=["Total dispo (kg/pers/an)","Population"])
TotalDispo=TotalDispo.groupby("Produit").sum()["TotGrDispoAlim"]
TotalDispo=TotalDispo.reset_index()
#Total dispo en grammes
```

```python
DispoProtPopu=pd.merge(DispoProt,data,on="Pays")
DispoProtPopu=DispoProtPopu.assign(TotProtDispo=conversionProt(DispoProtPopu["Protéine dispo(g/pers/jour)"],DispoProtPopu.Population))
ProtDispo=DispoProtPopu.drop(columns=["Protéine dispo(g/pers/jour)","Population"])
ProtDispo=ProtDispo.groupby("Produit").sum()["TotProtDispo"]
ProtDispo=ProtDispo.reset_index()
#Protéines totale dispo en grammes
```

```python
PourcentProteine=pd.merge(ProtDispo,TotalDispo,on="Produit")
```

```python
PourcentProteine=PourcentProteine.assign(Pourcentage=Calculfinal(PourcentProteine.TotProtDispo,PourcentProteine.TotGrDispoAlim))
```

```python
#PourcentProteine["RatioProteinePoids"]=(PourcentProteine["TotProtDispo"]/PourcentProteine["TotGrDispoAlim"])*100
```

```python
PourcentProteine=PourcentProteine.drop(columns=["TotProtDispo","TotGrDispoAlim"])
```

```python
PourcentProteine=PourcentProteine.dropna()
```
