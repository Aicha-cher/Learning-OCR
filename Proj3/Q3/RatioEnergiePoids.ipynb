```python
import pandas as pd
Popu2013=pd.read_csv("Popu2013.csv",index_col=0)
EnergiePoids=pd.read_csv("FAOSTAT_EnergiePoids.csv")
#FAOSTAT_EnergiePoids.csv contient la disponibilité alimentaire (pour les humains) en Kcal/personne/jour
#et en kilos/personne/an pour tous les produits et tous les pays. Année 2013
EnergiePoids=EnergiePoids.loc[EnergiePoids.Valeur !=0]
#Pour retirer toutes les valeures égales a 0
```

```python
Poids=EnergiePoids.groupby("Élément").get_group("Disponibilité alimentaire en quantité (kg/personne/an)")
```

```python
Energie=EnergiePoids.groupby("Élément").get_group("Disponibilité alimentaire (Kcal/personne/jour)")
```

```python
PoidsPopu=pd.merge(Poids,Popu2013,on="Pays")
PoidsPopu["Dispo alimentaire en kilos"]=PoidsPopu["Valeur"]*PoidsPopu["Population"]
PoidsClean=PoidsPopu.drop(columns=["Domaine","Élément","Année","Valeur","Population"])
#PoidsClean contient la disponibilité alimentaire en Kilos pour l'année 2013 pour l'ensemble de la population du pays
```

```python
EnergiePopu=pd.merge(Energie,Popu2013,on="Pays")
EnergiePopu["Dispo alimentaire en Kcal"]=EnergiePopu["Valeur"]*EnergiePopu["Population"]*365
EnergieClean=EnergiePopu.drop(columns=["Domaine","Élément","Année","Valeur","Population"])
#EnergieClean contient la disponibilité alimentaire en Kcal pour l'année 2013 pour l'ensemble de la population du pays
```

```python
RatioEnergiePoids=pd.merge(PoidsClean,EnergieClean,on=["Produit","Pays"])
RatioEnergiePoids["RatioEnergiePoids"]=RatioEnergiePoids["Dispo alimentaire en Kcal"]/RatioEnergiePoids["Dispo alimentaire en kilos"]
#Calcul et ajout de la colonne RatioEnergiePoids
```

```python
RatioEnergiePoids
#Pour tous les pays et tous les produits
```

```python
MoyenneEnergiePoids=RatioEnergiePoids.groupby("Produit").mean()["RatioEnergiePoids"]
#Moyenne pour tous les produits du ratio Kcal/Kg
```

```python
MoyenneEnergiePoids=pd.DataFrame(MoyenneEnergiePoids)
```

```python
MoyenneEnergiePoids=MoyenneEnergiePoids.reset_index()
```
