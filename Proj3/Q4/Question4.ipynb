```python
import pandas as pd
data=pd.read_csv("FAOSTAT_VegDispoInt.csv")
data=data.loc[data.Valeur !=0]
#Je retire les valeurs égales a 0 car elles sont indisponibles plutôt que nulles.
MoyenneEnergiePoids=pd.read_csv("MoyenneEnergiePoids.csv",index_col=0)
Popu2013=pd.read_csv("Popu2013.csv")

#FAOSTAT_VegDispoInt.csv ne contient que les éléments végétaux pour l'année 2013
#Disponibilité intérieure, Semences, Transformés et Autres Utilisations
```

```python
DispoInt=data.groupby("Élément").get_group("Disponibilité intérieure")
```

```python
DispoInt=DispoInt.rename(columns={"Valeur":"Milliers de tonnes"})
DispoInt=DispoInt.drop(columns=["Domaine","Élément","Année"])
DispoIntKilos=DispoInt.groupby("Produit").sum().reset_index()
DispoIntKilos["Kilos/an"]=DispoIntKilos["Milliers de tonnes"]*1000000
DispoIntKilos=DispoIntKilos.drop(columns="Milliers de tonnes")
```

```python
DispoIntKilos
```

```python
ApportVegetaux=pd.merge(DispoIntKilos,MoyenneEnergiePoids,on="Produit")
```

```python
ApportVegetaux["KcalDispoInt"]=ApportVegetaux["Kilos/an"]*ApportVegetaux["RatioEnergiePoids"]
```

```python
ApportVegetaux
#Réponse a la disponibilité intérieure mondiale exprimée en Kcal pour les produits végétaux
#La FAO recommande 2414 Kcal par jour en moyenne par personne.
#Ce qui fait 2414*365=881110 Kcal annuellement
```

```python
ApportVegetaux.sum()["KcalDispoInt"]/881110
#Résultat: 15 milliards de personnes pourraient être nourries chaque jour pendant une année
#Avec la totalité des produits végétaux SI elle était utilisée en intégralité pour l'alimentation humaine.
```

```python
AutreUtil=data[data.Élément != "Disponibilité intérieure"]
AutreUtil=AutreUtil.rename(columns={"Valeur":"Kilos"})
AutreUtil=AutreUtil.drop(columns=["Domaine","Élément","Année"])
AutreUtil=AutreUtil.groupby("Produit").sum().reset_index()
AutreUtil["Kilos"]=AutreUtil["Kilos"]*1000000
```

```python
ApportVegAutreUtil=pd.merge(AutreUtil,MoyenneEnergiePoids,on="Produit")
```

```python
ApportVegAutreUtil["KcalDispoInt"]=ApportVegAutreUtil["Kilos"]*ApportVegAutreUtil["RatioEnergiePoids"]
```

```python
(ApportVegetaux.sum()["KcalDispoInt"]-ApportVegAutreUtil.sum()["KcalDispoInt"])/881110
#Résultat: Presque 11 milliards d'humains pourraient être nourris chaque jour pendant un an
#si la production végétale, mis à part ses utilisations en Semence, 
#Traitement et Autres utilisation, était utilisée pour de la nourriture
```

```python
(ApportVegAutreUtil.sum()["KcalDispoInt"]/ApportVegetaux.sum()["KcalDispoInt"])*100
#Juste pour le fun, le pourcentage de produits végétaux utilisés en Semence, 
#Traitement et Autres utilisation, sur les produits végétaux disponibles totaux
```
