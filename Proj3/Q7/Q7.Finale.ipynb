```python
import pandas as pd
#Un fichier pour les céréales partant pour l'alimentation animale
data=pd.read_csv("FAOSTAT_CerealesPourAnimaux.csv")
#Un fichier pour les produits animaux déstinés à l'alimentation humaine
df=pd.read_csv("FAOSTAT_AnimAlim.csv")
#Les deux tables de conversion
RatioProteinePoids=pd.read_csv("RatioProteinePoids.csv",index_col=0)
MoyenneEnergiePoids=pd.read_csv("MoyenneEnergiePoids.csv",index_col=0)
```

```python
data["Kilos"]=data["Valeur"]*1000000
CerealesPourAnimaux2013=data.drop(columns=["Domaine","Pays","Élément","Année","Valeur"])
#Simplification de la table (conversion de milliers de tonnes en kilos)
```

```python
df["Kilos"]=df["Valeur"]*6997326000
AnimauxPourHumains2013=df.drop(columns=["Domaine","Pays","Élément","Année","Valeur"])
#Simplification de la table (conversion de kg/personne/an en kilos en multipliant par la population totale)
```

```python
CerealesPourAnimaux2013=pd.merge(CerealesPourAnimaux2013,RatioProteinePoids,on="Produit")
CerealesPourAnimaux2013=pd.merge(CerealesPourAnimaux2013,MoyenneEnergiePoids,on="Produit")
CerealesPourAnimaux2013["Grammes de Protéines"]=CerealesPourAnimaux2013.Kilos*(CerealesPourAnimaux2013.Pourcentage/100)
CerealesPourAnimaux2013["Kcalories"]=CerealesPourAnimaux2013.Kilos*CerealesPourAnimaux2013.RatioEnergiePoids
CerealesPourAnimaux2013=CerealesPourAnimaux2013.drop(columns=["Pourcentage","RatioEnergiePoids"])
#Fusion avec les deux tables de conversion, créations des deux nouvelles colonnes, puis élimination des colonnes inutiles
```

```python
AnimauxPourHumains2013=pd.merge(AnimauxPourHumains2013,RatioProteinePoids,on="Produit")
AnimauxPourHumains2013=pd.merge(AnimauxPourHumains2013,MoyenneEnergiePoids,on="Produit")
AnimauxPourHumains2013["Grammes de Protéines"]=AnimauxPourHumains2013.Kilos*(AnimauxPourHumains2013.Pourcentage/100)
AnimauxPourHumains2013["Kcalories"]=AnimauxPourHumains2013.Kilos*AnimauxPourHumains2013.RatioEnergiePoids
AnimauxPourHumains2013=AnimauxPourHumains2013.drop(columns=["Pourcentage","RatioEnergiePoids"])
#Fusion avec les deux tables de conversion, créations des deux nouvelles colonnes, puis élimination des colonnes inutiles
```

```python
CerealesPourAnimaux2013.sum()["Kilos"]/AnimauxPourHumains2013.sum()["Kilos"]
#Valeur du ratio "Disponibilité de céréales destinée à l'alimentation animale/Disponibilité animale
#destinée à l'alimentation humaine" en Kilos.
```

```python
CerealesPourAnimaux2013.sum()["Grammes de Protéines"]/AnimauxPourHumains2013.sum()["Grammes de Protéines"]
#Valeur du ratio "Disponibilité de céréales destinée à l'alimentation animale/Disponibilité animale
#destinée à l'alimentation humaine" en protéines.
```

```python
CerealesPourAnimaux2013.sum()["Kcalories"]/AnimauxPourHumains2013.sum()["Kcalories"]
#Valeur du ratio "Disponibilité de céréales destinée à l'alimentation animale/Disponibilité animale
#destinée à l'alimentation humaine" en Kcalories.
```
