```python
import pandas as pd
df=pd.read_csv("FAOSTAT_DispoInterieureProdAnim.csv")
RatioProtPoids=pd.read_csv("RatioProteinePoids.csv",index_col=0)
EnergiePoids=pd.read_csv("MoyenneEnergiePoids.csv",index_col=0)
```

```python
df["Kilos"]=df["Valeur"]*1000000
```

```python
DispoInterieureProduitsAnim=df.drop(columns=["Domaine","Pays","Élément","Année","Valeur"])
```

```python
DispoInterieureProduitsAnim=pd.merge(DispoInterieureProduitsAnim,RatioProtPoids,on="Produit")
```

```python
DispoInterieureProduitsAnim=pd.merge(DispoInterieureProduitsAnim,EnergiePoids,on="Produit")
```

```python
DispoInterieureProduitsAnim["Gramme de Protéines"]=DispoInterieureProduitsAnim["Kilos"]*1000*(DispoInterieureProduitsAnim["Pourcentage"]/100)
```

```python
DispoInterieureProduitsAnim["Kcal"]=DispoInterieureProduitsAnim["Kilos"]*DispoInterieureProduitsAnim["RatioEnergiePoids"]
```

```python
DispoInterieureProduitsAnim=DispoInterieureProduitsAnim.drop(columns=["Pourcentage","RatioEnergiePoids"])
```
